# Analyzing Salaries Post Undergrad by College Type 

```python
#import libraries

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
#import cvs file

location = "salaries-by-college-type.csv"
df = pd.read_csv(location)
df.head()
```

```python
df.count()
```

```python
df.dtypes
```

```python
#rename columns
df.rename(columns={'School Name': 'SchoolName','School Type': 'SType', 'Starting Median Salary': 'StartingMedianSalary','Mid-Career Median Salary':'MidCareerMedianSalary', 'Mid-Career 25th Percentile Salary': 'MidCareer25Salary', 'Mid-Career 75th Percentile Salary' : 'MidCareer75Salary'}, inplace=True)
```

```python
df.head()
```

In order to convert the date types from an object to a integer you have to remove the "$" and ","

```python
df['StartingMedianSalary'] = df['StartingMedianSalary'].apply(lambda x:x.strip('$'))
```

```python
df['StartingMedianSalary'] = df['StartingMedianSalary'].apply(lambda x:x.replace(',', ''))
```

```python
df['MidCareerMedianSalary'] = df['MidCareerMedianSalary'].apply(lambda x:x.strip('$'))
```

```python
df['MidCareerMedianSalary'] = df['MidCareerMedianSalary'].apply(lambda x:x.replace(',', ''))
```

```python
df['MidCareer25Salary'] = df['MidCareer25Salary'].apply(lambda x:x.strip('$'))
```

```python
df['MidCareer25Salary'] = df['MidCareer25Salary'].apply(lambda x:x.replace(',', ''))
```

```python
df['MidCareer75Salary'] = df['MidCareer75Salary'].apply(lambda x:x.strip('$'))
```

```python
df['MidCareer75Salary'] = df['MidCareer75Salary'].apply(lambda x:x.replace(',', ''))
```

Convert data type from an object to a float then from a float to a integer

```python
df['StartingMedianSalary'] = df['StartingMedianSalary'].astype(float)
```

```python
df['StartingMedianSalary'] = df['StartingMedianSalary'].astype(int)
```

```python
df['MidCareerMedianSalary'] = df['MidCareerMedianSalary'].astype(float)
```

```python
df['MidCareerMedianSalary'] = df['MidCareerMedianSalary'].astype(int)
```

```python
df['MidCareer25Salary'] = df['MidCareer25Salary'].astype(float)
```

```python
df['MidCareer25Salary'] = df['MidCareer25Salary'].astype(int)
```

```python
df['MidCareer75Salary'] = df['MidCareer75Salary'].astype(float)
```

```python
df['MidCareer75Salary'] = df['MidCareer75Salary'].astype(int)
```

```python
df.groupby('SType').mean()
```

```python
grouped = df.groupby('S', sort=False).mean()
grouped.sort_index(ascending = False)
```

```python
result = smf.ols('StartingMedianSalary ~ SType', data = df).fit()
result.summary()
```

```python
result = smf.ols('MidCareerMedianSalary	 ~ SType', data = df).fit()
result.summary()
```

```python
result = smf.ols('MidCareer75Salary ~ SType', data = df).fit()
result.summary()
```

```python
result = smf.ols('MidCareer25Salary ~ SType', data = df).fit()
result.summary()
```
